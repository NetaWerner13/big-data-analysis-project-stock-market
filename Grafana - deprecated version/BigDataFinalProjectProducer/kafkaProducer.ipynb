```python
# required imports
from confluent_kafka import Producer
import time
import csv
import json
```

```python
##### KAFKA PRODUCER #####
# Initialize Kafka producer and program variables
file_path = "mock_transaction_data.csv"
topic = "transactions"
server = 'localhost:9092'
chunk_size = 100
producer = Producer({'bootstrap.servers': server})
```

```python
# read lines from the file
def read_lines(csv_file, batch_size=100):
    with open(csv_file, mode='r') as file:
        csv_reader = csv.DictReader(file)
        batch = []
        for row in csv_reader:
            batch.append(row)
            if len(batch) == batch_size:
                send_batch(batch)
                batch = []      
        # Send any remaining rows
        if batch:
            send_batch(batch)
```

```python
# send batch of messages to Kafka
def send_batch(batch):
    for row in batch:
        message = json.dumps(row).encode('utf-8')
        producer.produce(topic, message)
    producer.flush()
    print(f"Sent {len(batch)} messages to Kafka")
    time.sleep(1)
```

```python
if __name__ == '__main__':
    read_lines(file_path, chunk_size)
```
